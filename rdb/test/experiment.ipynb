```python
import pandas as pd
from pathlib import Path

data_dir = Path("data")
files = sorted(f.name for f in data_dir.glob("test-data-2025-*.csv"))

total_seconds = 0
unique_vehicle_ids = set()
unique_trip_ids = set()
unique_route_ids = set()
total_observations = 0

for f in files:
    df = pd.read_csv(data_dir / f)
    total_seconds += df["duration"].sum()
    unique_vehicle_ids.update(df["vehicle_id"])
    unique_trip_ids.update(df["trip_id"])
    unique_route_ids.update(df["route_id"])
    total_observations += len(df)

# convert seconds to years (365 days/year)
seconds_per_year = 365 * 24 * 3600
total_years = int(total_seconds / seconds_per_year)

print(f"Total duration (seconds): {total_seconds:,}")
print(f"Total duration (years):   {total_years}")
print(f"Unique vehicle IDs:       {len(unique_vehicle_ids):,}")
print(f"Unique trip IDs:          {len(unique_trip_ids):,}")
print(f"Unique route IDs:         {len(unique_route_ids):,}")
print(f"Total observations:       {total_observations:,}")
```
